```python
import json
from os.path import join as pjoin
import nltk
```

```python
def data_from_json(filename):
    data = []
    with open(filename) as data_file:
        for l in data_file:
            data.append(json.loads(l))
    return data
```

```python
train_file_name = pjoin("queries", "freebuild.json")
data = data_from_json(train_file_name)
```
